```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Raw_Skills_Dataset.csv');
de = pd.read_csv('/content/Example_Technical_Skills.csv');
print(df)
print(de)
```

```python
# Checking if any duplicate row exist in Raw_Skills_Dataset and printing it if it exists. 
dups = df.duplicated()
print(dups.any())
print(df[dups])
```

```python
# Deleting duplicate rows from Raw_Skills_Dataset
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
# Checking if any duplicate row exist in Example_Technical_Skills and printing it if it exists. 
dups2 = de.duplicated()
print(dups2.any())
print(de[dups2])
```

```python
# Deleting duplicate rows from Example_Technical_Skills
de.drop_duplicates(inplace=True)
print(de.shape)
```

```python
# Checking whether the column "RAW DATA" in Raw_Skills_Dataset contains null values or not 
bool_series = pd.isnull(df["RAW DATA"])
df[bool_series]
```

```python
# Checking whether the column "Technology Skills" in Example_Technical_Skills contains null values or not 
bool_series2 = pd.isnull(de["Technology Skills"])
df[bool_series]
```

```python
# Converting Example_Technical_Skills in to numpy array(2D)
numpy_array = de.to_numpy()
numpy_array.shape
```

```python
# Converting numpy array into 1D array
result = numpy_array.flatten()
result
```

```python
# selecting rows based on condition
rslt_df = df[df['RAW DATA'].isin(result)]
rslt_df
```
